# Data Cleaning and Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
plt.figure(figsize = (20,20))
sbn.heatmap(df.corr(numeric_only = True),annot=True)
```

# Feature Engineering

```python
df.columns
```

```python
df['age_thalach'] = df['age'] * df['thalach']
```

```python
df['oldpeak_slope'] = df['oldpeak'] * df['slope']
```

```python
df['ca_thal'] = df['ca'] * df['thal']
```

```python
df['cp_exang'] = df['cp'] * df['exang']
```

```python
df['age_chol'] = df['age'] * df['chol']
```

```python
df['bp_age'] = df['trestbps'] * df['age']
```

```python
df['heart_risk_score'] = (
    df['exang'] + 
    df['oldpeak'] + 
    df['ca'] + 
    df['thal']
)
```

```python
df['high_oldpeak'] = (df['oldpeak'] > 2).astype(int)
df['low_thalach'] = (df['thalach'] < 120).astype(int)
df['high_ca'] = (df['ca'] >= 2).astype(int)
```

```python
plt.figure(figsize = (20,20))
sbn.heatmap(df.corr(numeric_only = True),annot=True)
```

```python
df.columns
```

```python
features = ['age', 'sex','cp','chol','restecg','exang','thalach','oldpeak','slope','ca','thal','heart_risk_score',
    'ca_thal','age_chol',
    'oldpeak_slope','target']
```

```python
fianl_heart_df = df[features]
```

```python
fianl_heart_df.shape
```

```python
plt.figure(figsize = (20,20))
sbn.heatmap(fianl_heart_df.corr(numeric_only = True),annot=True)
```

# model training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_predict
import warnings

warnings.filterwarnings('ignore')
```

```python
scaler = StandardScaler()
```

```python
X = fianl_heart_df.drop(labels = 'target', axis = 1)
y = fianl_heart_df['target']
scaled_data = scaler.fit_transform(X)
X_scaled = pd.DataFrame(scaled_data, columns=X.columns)
```

```python
X_Strain, X_Stest, y_Strain, y_Stest = train_test_split(X_scaled, y, test_size=0.20, random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# logistic Regression

```python
LR_model = LogisticRegression(max_iter = 1000)
```

```python
LR_model.fit(X_Strain,y_train)
```

```python
LR_model.fit(X_train,y_train)
```

```python
y_predict = LR_model.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
print("Accuracy:", accuracy_score(y_test, y_predict))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_predict))
print("\nClassification Report:\n", classification_report(y_test, y_predict))
```

```python
y_predictS = LR_model.predict(X_Stest)
```

```python
print("Accuracy:", accuracy_score(y_test, y_predictS))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_predictS))
print("\nClassification Report:\n", classification_report(y_test, y_predictS))
```

```python
y_predictM = cross_val_predict(LR_model, X, y, cv=5)
```

```python
print("Accuracy:", accuracy_score(y, y_predictM))
print("\nConfusion Matrix:\n", confusion_matrix(y, y_predictM))
print("\nClassification Report:\n", classification_report(y, y_predictM))
```

```python
y_predictSc = cross_val_predict(LR_model, X_scaled, y, cv=5)
```

```python
print("Accuracy:", accuracy_score(y, y_predictSc))
print("\nConfusion Matrix:\n", confusion_matrix(y, y_predictSc))
print("\nClassification Report:\n", classification_report(y, y_predictSc))
```
