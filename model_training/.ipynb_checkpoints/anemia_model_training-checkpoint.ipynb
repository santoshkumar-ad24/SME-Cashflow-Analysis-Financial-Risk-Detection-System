# feature Engineering and extraction

```python
df = copied_anemia_data.copy()
```

```python
df.head()
```

```python
df['low_hb'] = (df['Hemoglobin'] < 12).astype(int)
```

```python
df['hb_severity'] = pd.cut(
    df['Hemoglobin'],
    bins=[0, 8, 11, 13, 20],
    labels=[3, 2, 1, 0]
)
```

```python
df['rbc_index'] = df['MCH'] + df['MCHC']
```

```python
df['hb_mch'] = df['Hemoglobin'] * df['MCH']
df['hb_mchc'] = df['Hemoglobin'] * df['MCHC']
```

```python
df['mch_mcv_ratio'] = df['MCH'] / (df['MCV'] + 1)
df['hb_mcv_ratio'] = df['Hemoglobin'] / (df['MCV'] + 1)
```

```python
df['hb_gender'] = df['Hemoglobin'] * df['Gender']
```

```python
df.head()
```

```python
plt.figure(figsize = (10,10))
sea.heatmap(df.corr(numeric_only = True),annot=True)
```

```python
df.columns
```

```python
final_anemia_df = df[[
    'Hemoglobin',   
    'Gender',
    'rbc_index',
    'mch_mcv_ratio',
    'hb_mcv_ratio', 'MCH', 'MCHC', 'MCV', 'Result'
]]
```

```python
plt.figure(figsize = (10,10))
sea.heatmap(final_anemia_df.corr(numeric_only = True),annot=True)
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

# model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_predict, StratifiedKFold
```

```python
X = final_anemia_df.drop(labels = 'Result', axis = 1)
y = final_anemia_df['Result']
scaled_data = scaler.fit_transform(X)
X_scaled = pd.DataFrame(scaled_data, columns=X.columns)
X_scaled.head()
```

```python
LR_model = LogisticRegression()
y_predict = cross_val_predict(LR_model, X, y, cv=5)
y_predictS = cross_val_predict(LR_model, X_scaled, y, cv=5)
```

```python
LR_model.fit(X,y)
```

```python

print("Accuracy:", accuracy_score(y, y_predict))
print("\nConfusion Matrix:\n", confusion_matrix(y, y_predict))
print("\nClassification Report:\n", classification_report(y, y_predict))
```

```python

print("Accuracy:", accuracy_score(y, y_predictS))
print("\nConfusion Matrix:\n", confusion_matrix(y, y_predictS))
print("\nClassification Report:\n", classification_report(y, y_predictS))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python


pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
y_predict_RF = cross_val_predict(pipeline, X, y, cv=5)
```

```python
              
print("Accuracy:", accuracy_score(y, y_predict_RF))
print("\nConfusion Matrix:\n", confusion_matrix(y, y_predict_RF))
print("\nClassification Report:\n", classification_report(y, y_predict_RF))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_trainS, X_testS, y_trainS, y_testS = train_test_split(X_scaled, y, test_size=0.20, random_state=42)
```

```python
LR_model = LogisticRegression()
```

```python
LR_model.fit(X_train, y_train)
y_predict = LR_model.predict(X_test)
LR_model.fit(X_trainS, y_trainS)
```

```python
y_predictS = LR_model.predict(X_testS)
```

```python
print("Accuracy:", accuracy_score(y_test, y_predict))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_predict))
print("\nClassification Report:\n", classification_report(y_test, y_predict))
```

```python
print("Accuracy:", accuracy_score(y_test, y_predictS))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_predictS))
print("\nClassification Report:\n", classification_report(y_test, y_predictS))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python

pipeline_KNN = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors = 7, algorithm = 'ball_tree', metric =  'manhattan'))
])
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
pipeline_KNN = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])
```

```python
param_grid = {'knn__n_neighbors': range(5, 25),
             'knn__algorithm': ['ball_tree', 'kd_tree', 'brute'],
              'knn__metric': ['euclidean', 'manhattan']
             } 

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(pipeline_KNN, param_grid, cv=cv, scoring='accuracy')
grid.fit(X, y)

print("Best n_neighbors:", grid.best_params_['knn__n_neighbors'])
print("Best algorithm:", grid.best_params_['knn__algorithm'])
print("Best metric:", grid.best_params_['knn__metric'])

print("Best cross-validation accuracy:", grid.best_score_)
```
